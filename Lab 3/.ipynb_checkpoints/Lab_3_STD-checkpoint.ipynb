```python
def myMedian(img):
    window_wh = 3
    #edge_x = edge_y = floor(window_wh / 2)
    height, width = img.shape
    med = np.zeros_like(img)
    for x in range(1, width - 1):
        for y in range(1, height - 1):
            #shifts = (0,0), (0,1), (0,-1), (1,0), (1,1), (1,-1), (-1,0), (-1,1), (-1,-1)
            #for shift in shifts:
            #    nbs.append(img[y+shift[1],x+shift[0]])
            nbs = img[y:y+3,x:x+3]
            nbs = nbs.flatten()
            nbs.sort()
            med[y,x] = nbs[int(nbs.size / 2)]
    return med

img = rgb2gray(io.imread('pout.tif'))
med = np.zeros_like(img)
x = y = 2
nbs = img[y:y+3,x:x+3]
nbs = nbs.flatten()
print(nbs.shape)
nbs.sort()
med[y,x] = nbs[4]
res = myMedian(rgb2gray(io.imread('pout.tif')))
show_images((res,))
```

```python
img = rgb2gray(io.imread('pout.tif'))
noisyImg = random_noise(img, mode='s&p', amount=0.05)
med = median(noisyImg)
medImpl = myMedian(noisyImg)

sigmas = np.arange(0,4,0.4)#0.2, 0.5, 2, 8
print(sigmas)
gaussians = ([],[])
for sigma in sigmas:
    gaussians[0].append(gaussian(noisyImg, sigma=sigma))
    gaussians[1].append(str(sigma))

show_images([img, noisyImg, med, medImpl], ['original', 'noised', 'med', 'myMed'])
show_images(gaussians[0],gaussians[1])
```
